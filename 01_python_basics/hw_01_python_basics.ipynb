```python
import re
from typing import Any
def result_type(expr: str) -> str:
  """
  Simple function for calculating the result of expression
  and displaying it's type.

  Args:
    expr (str): The expression for analize ande calculating

  Returns:
    str: The result of expression and name of it's type
  """

  oper_dict: dict = {
      "*" : "множення",
      "+" : "додавання",
      "-" : "віднімання",
      "/" : "ділення",
      "//" : "цілочисельне ділення",
      "%" : "залишок від ділення",
      "**" : "піднесення до степеня",
      ">" : "порівняння (більше)",
      ">=" : "порівняння (більше або дорівнює)",
      "<" : "порівняння (менше)",
      "<=" : "порівняння (менше або дорівнює)",
      "==" : "порівняння (дорівнює)",
      "!=" : "порівняння (не дорівнює)",
      }
  used_oper_set: set = set()
  oper_sort: list = sorted(oper_dict.keys(), key=len, reverse=True)
  split_expr: list = expr.split()
  for op in oper_sort:
    pattern: str = re.escape(op)
    match: re.Match|None = re.search(pattern, expr)
    if match:
      used_oper_set.add(oper_dict[op])
  result: Any = eval(expr) #дуже небезпечна штука, краще не використовувати у проді
  return (f"Виконуючи {", ".join([i for i in used_oper_set])}, в результаті отримали "
    f"'{result}' з типом даних {type(result).__name__}.")

print(result_type(input('Будь ласка введіть арифметичну операцію: ')))


```
