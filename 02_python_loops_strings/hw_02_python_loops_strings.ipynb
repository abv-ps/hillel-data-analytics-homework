Додатково виведіть скільки слів орієнтовно знає студент на певному рівні (користуйтеся словником number_of_words).

Шкала від 0 до 100:

```

```

Приклад:

```
Input: 67
Output: Студент має рівень англійської С1 (61-80) балів, та орієнтовно знає 8000 слів.
```




````python
def lvl_of_eng(test_result: str) -> str:
  """
  Simple function for calculating the level of english
  and displaying approximate student's vocabulary size.

  Args:
    test_result (str): The student's test result for analize and calculating

  Returns:
    str: The student's level of english and approximate vocabulary size
  """
  lvl_dict: dict = {
      0:'A1',
      20:'A2',
      40:'B1',
      60:'B2',
      80:'C1',
      100:'C2'
  }
  number_of_words: dict =  {
      'A1': 500,
      'A2': 1000,
      'B1': 2000,
      'B2': 4000,
      'C1': 8000,
      'C2': 16000
  }
  while not test_result.isdigit() or test_result == 0 or int(test_result) > 100:
    test_result: str = input("Ви ввели некоректне значення. Введіть будь ласка коректний результат тестування від 0 до 100: ")
  int_res: int = int(test_result)
  for i in lvl_dict.keys():
    if int_res <= int(i):
      lvl_st: str = lvl_dict[i]
      if i==0:
        score: str = 0
      else:
        score: str = str((max(0, i-1)//20)*20+1) + '-' + str((max(0, i-1)//20+1)*20)
      break
  return (f"Рівень англійської студента '{lvl_st}' {'рівний ' if i==0  else 'знаходиться у діапазоні '}{score} "
        f"балів, а словарний запас орієнтовно складає {number_of_words[lvl_st]} слів")
print(lvl_of_eng(input("Введіть будь ласка результат тестування від 0 до 100: ")))
````

Напишіть алгоритм, що виводитиме на кожній ітерації наступні показники:

(для значень  Х = 0;  Y = 100; Z = 10)

* 1 місяць. Зібрано 10$. Треба зібрати ще 90€.

* 2 місяць. Зібрано 20$. Треба зібрати ще 80€.

* ....

І так аж поки ціль не буде досягнута (включно з останнім місяцем, коли зібрати треба вже 0).



````python
in_account = 17
goal = 137
monthly_income = 15
month = 0
while not in_account == goal:
  in_account += monthly_income
  left_to_goal = goal-in_account
  month += 1
  print (f"{month}-й місяць. Зібрано {monthly_income*month}$. Залишилось зібрати ще {left_to_goal}$.")
````
