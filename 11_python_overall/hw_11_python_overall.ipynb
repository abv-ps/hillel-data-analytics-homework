**1. Як можна обробити пропущені значення у DataFrame за допомогою pandas? Наведіть 2 приклади.**

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(
    dict(
        age=[5, 6, np.nan],
        born=[
            pd.NaT,
            pd.Timestamp("1939-05-27"),
            pd.Timestamp("1940-04-25"),
        ],
        name=["Alfred", "Batman", ""],
        toy=[None, "Batmobile", "Joker"],
    )
)
df
```

Створюємо датафрейм з відповідними значеннями пропусків

```python
df.replace(r'^\s*$', pd.NaT, regex=True).isna()
```

Бачимо, в яких клітинках ці значення присутні. Пусті значення не відображаються цією функцією, тому їх шукаємо регексом через реплейс.

```python
df.replace(r'^\s*$', pd.NaT, regex=True).dropna(subset=["name", "toy"])
```

Отримуємо відображення копії датафрейму, яка показує тільки fullfilled рядки.

```python
values = {'age': df['age'].mean(), 'born': 'SomeDay', 'name': 'SomeName', 'toy': 'SomeToy'}
df.replace(r'^\s*$', pd.NaT, regex=True).fillna(value=values)
```

Дивимось на копію датафрейму, заповнену вигаданими даними.

**2. Як об'єднати два набори даних у pandas за спільною колонкою?**

---



```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
print(df1)
print('')
print(df2)
```

Згідно документацйії створюємо два датафрейми.

```python
df1.merge(df2, left_on='lkey', right_on='rkey',
          suffixes=('_left', '_right'))
```

Отримуємо об'єднаний набір із зрозумілими назвами стовпчиків.

**4. Яка різниця між mutable та immutable змінними у Python? Наведіть приклад кожного типу.**

При створенні об'єкта їй присвоюється адреса в пам'яті, якщо ця адреса лишається незмінною при коригуванні об'єкта - він mutable (список, словник, множина). При зміні immutable (int, float, str, tuple) об'єкта створюється новий об'єкт з іншою адресою в пам'яті.

```python
a_list = ["foo", 2, [4, 5]]
print(f'a_list: {a_list}\na_list_id: {id(a_list)}')
a_list[2] = (3, 4)
print(f'a_list: {a_list}\na_list_id: {id(a_list)}')
```

```python
i_int = 1
print(f'i_int: {i_int}\ni_int_id: {id(i_int)}')
i_int += 1
print(f'\ni_int: {i_int}\ni_int_id: {id(i_int)}')

i_float = 1.0
print(f'\ni_float: {i_float}\ni_float_id: {id(i_float)}')
i_float += 1.0
print(f'\ni_float: {i_float}\ni_float_id: {id(i_float)}')

i_str = 'a'
print(f'\ni_str: {i_str}\ni_str_id: {id(i_str)}')
i_str = 'a'*2
print(f'\ni_str: {i_str}\ni_str_id: {id(i_str)}')

i_tupple = (i_int, i_float, i_str)
print(f'\ni_tupple: {i_tupple}\ni_tupple_id: {id(i_tupple)}')
i_tupple = i_tupple*2
print(f'\ni_tupple: {i_tupple}\ni_tupple_id: {id(i_tupple)}')
```

**8. Що таке groupby() у pandas? Наведіть приклад його використання.**

```python
df = pd.DataFrame({
                  "key1" : ["a", "a", None, "b", "b", "a", None],
                  "key2" : pd.Series([1, 2, 1, 2, 1, None, 1], dtype="Int64"),
                  "data1" : np.random.standard_normal(7),
                  "data2" : np.random.standard_normal(7)
                  })
df
```

```python
grouped = df["data1"].groupby(df["key1"]) #групуємо дані по ключу key1
grouped.mean() # відображаємо середнє значення згрупованих результатів
```
