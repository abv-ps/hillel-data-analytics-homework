### 🟡 1. (20 балів) Списки...
Створіть випадковий список на 9 елементів. Виконайте з ним наступні дії:

* додати в кінець списку -1
* додати в кінець списку елементи списку [-2, -8, -5, -7]
* третім елементом додати 9
* порахувати кількість трьох випадкових чисел (з списку) в списку
* визначити який індекс в списку має елемент -5
* перевірити чи є в списку -2, -10
* відсортувати список
* розвернути відсортований список

````python
import random
from itertools import repeat
def list_operations(lst: list) -> None:
  """
  Simple function for performing operations based on a given list.

  Args:
    lst (list): The given list for performing operations

  Returns:
    None: Display the description and result of the operations performed
          on the given list
  """
  print(f"Початковий список: {lst}")
  operations = [
      ("Додали в кінець списку один елемент", "append", (-1,), {}),
      ("Додали до списку елементи списку", "extend", ([-2, -8, -5, -7],), {}),
      ("Вставили '9' на 3 позицію один елемент", "insert", (2, 9), {}),
      (("Кількість трьох випадкових чисел", "у списку", "складає відповідно"), "count", (), {}),
      ("Індекс елементу '-5' у списку, його індекс ", "index", (-5,), {}),
      (("Елемент '-2' присутній у списку стільки разів", "Елемент '-2' не присутній у списку"), "count", (-2,), {}),
      (("Елемент '-10' присутній у списку стільки разів", "Елемент '-10' не присутній у списку"),"count", (-10,), {}),
      ("Відсортований список", "sort", (), {}),
      ("Обернений відсортований список", "sort", (), {"reverse":True})
  ]
  for desc, oper, args, kwargs in operations:
    if oper == "count" and "випадкових" in desc[0]:
      chosen_numbers = [random.choice(lst) for _ in range(3)]
      action_count = [getattr(lst, oper)(num) for num in chosen_numbers]
      print(f"{desc[0]} {chosen_numbers} {desc[1]} {lst} {desc[2]} {action_count}")
    else:
      action: int|None = getattr(lst, oper)(*args, **kwargs)
      if isinstance(desc, (tuple)):
        current_desc = desc[0] if action else desc[1]
      else:
        current_desc = desc
      print(f"{current_desc}: {action}" if action is not None else f"{current_desc}: {lst}")
  return
some_list = random.sample(range(1,10), 9)
list_operations(some_list)
````

## 🟡 6. (20 балів) Створіть об'єкт Counter з списку випадкових 100 чисел.

* Виведіть 2 об'єкти, що зустрічаються найчастіше і по скільки разів вони зустрічаються
* Видаліть з Counter інформацію про елемент, що зустрічається найчастіше
* Створіть ще один список на 10 чисел і оновіть об'єкт Counter цим новим списком.

````python
def get_suffix(num: int) -> str:
  if 11 <= num % 100 <= 14:
    return "разів"
  if num % 10 == 1:
    return "раз"
  if num % 10 <=4:
    return "рази"
  return "разів"

from collections import Counter
counter_list = Counter([random.randint(1,100) for _ in range(100)])
print('Найпоширеніші елементи у списку:')
for el, cnt in counter_list.most_common(2):
  print(f'{el} зустрічається {cnt} {get_suffix(cnt)}')
print(f'Видаляємо найпоширеніший елемент {counter_list.most_common(1)[0]}')
del counter_list[counter_list.most_common(1)[0][0]]
print(f'Тепер найпоширеніший елемент {counter_list.most_common(1)[0]}')
new_lst = [random.randint(1,100) for _ in range(10)]
print(f'Додаємо до лічильника новий список {new_lst}')
counter_list.update(new_lst)
print(f'Тепер найпоширеніший елемент {counter_list.most_common(1)[0]}')
````

````python
import random
currency_codes = {840:"USD", 978:"EUR", 980:"UAH"}
currency_symbols = {"UAH":"₴", "EUR":"€", "USD":"$"}
exchange_rate = {"UAH":1.0, "EUR":50.5, "USD":43.3}
def lvl_of_nesting(data: dict|list) -> dict:
  """
  Function for displaying information about dictionary key names
  and their nesting levels.

  Args:
    data (dict|list): The given dictionary or list for collecting information

  Returns:
    Dict: Display the description and result of the operations performed
          on the given dictionary or list
  """
  def recursive_nesting(cur_data: dict|list,
                        nested_list: list|None=None,
                        id_key: str|None=None) -> dict:
    if nested_list is None:
      nested_list = []
    cur_id_key = id_key
    if isinstance(cur_data, dict):
      for key, value in cur_data.items():
        cur_id_key = cur_data.get('id', id_key)
        nested_list.append({
                "id_key": cur_id_key,
                "key": key,
                "value": value
            })
        recursive_nesting(value, nested_list, cur_id_key)
    if isinstance(cur_data, list):
      for item in cur_data:
        recursive_nesting(item, nested_list, cur_id_key)
    return nested_list
  return recursive_nesting(data)

list_of_data = lvl_of_nesting(user_data)

unique_types = set()

maskedPan_type = {}

group_by_id_dict = {}

unique_currency_balance = {}

currency_count = {}

positive_credit_limit = []
simplified_account = []


for item in list_of_data:
  id_key, key, value = item['id_key'], item['key'], item['value']
  if not id_key is None:
    if id_key not in group_by_id_dict:
      group_by_id_dict[id_key] = {}
    group_by_id_dict[id_key][key]=value

i_keys = ['type', 'currencyCode', 'balance', 'maskedPan', 'creditLimit']


for item in group_by_id_dict:
  i_type, i_cur_code, i_bal, i_mPan, i_cLim = (
      group_by_id_dict[item].get(key) for key in i_keys
  )

  smpl_info = {key:group_by_id_dict[item].get(key) for key in i_keys[:4]}
  simplified_account.append(smpl_info)

  unique_types.add(i_type)
  currency_name = currency_codes.get(i_cur_code)
  unique_currency_balance[currency_name] = (
      unique_currency_balance.get(currency_name, 0)
  ) + i_bal
  currency_count[currency_name] = (
      currency_count.get(currency_name, 0)
      ) + 1
  for pan in i_mPan:
    maskedPan_type[pan] = (i_type, i_cur_code, i_bal)
  if i_cLim > 0:
    positive_credit_limit.append(item)





print(f"[1] Перелік унікальних типів рахунків:\n {'\n'.join(unique_types)}")
balance = '\n'.join([f'{cur}: {bal/ 100:.2f}{currency_symbols.get(cur, cur)}' for cur, bal in unique_currency_balance.items()])
print(f"\n[2] Сума балансів по кожній валюті:\n{balance}")
card_type = '\n'.join([f"Карта '{pan}' має тип '{type[0]}'" for pan, type in maskedPan_type.items()])
print(f"\n[3]\n", card_type)
print(f"\n[4] Рахунки з кредитним лімітом: '{'\', \''.join(positive_credit_limit)}'")
simple_acc ="\n".join([
    ", ".join([f"'{k}': '{currency_codes.get(v,v)}'" for k, v in list(item.items())[:3]])
      for item in simplified_account
    ])
print(f"\n[5] Основна інформація по рахункам аккаунта:\n{simple_acc}")
quantity = '\n'.join([f'{v}: {q} шт.' for v, q in currency_count.items()])
print(f"\n[6] Кількість рахунків по кожній валюті:\n{quantity}")
random_acc = random.choice(list(maskedPan_type.items()))
pan = random_acc[0][-8:]
type, cur, bal = random_acc[1]
print(f"\n[7] Картка {pan} ({type}, {currency_codes.get(cur,cur)}) має баланс "
      f"{bal/ 100:.2f}{currency_symbols.get(currency_codes.get(cur, cur))}"
  )
total_uah_balance = sum(unique_currency_balance[cur]*exchange_rate[cur]
                        for cur in unique_currency_balance)
print(f"\n[8] Загальний баланс у гривневому еквіваленті {total_uah_balance/ 100:.2f}{currency_symbols.get("UAH")}")
````
