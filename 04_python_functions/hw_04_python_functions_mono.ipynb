# Завдання

## 🔹 1. get_currency_code(currency: str) -> int

Призначення:
* Перетворює 3-літерний код валюти у числовий код Monobank.


Приклад:

```
get_currency_code("USD") → 840
get_currency_code("eur") → 978
get_currency_code("GBP") → None
```



````python
def get_currency_code(currency: str) -> int|None:
    """
    Приймає 3-літерний код валюти (наприклад "USD", "EUR", "UAH").
    Повертає числовий код валюти:
        980 — UAH
        840 — USD
        978 — EUR
    Якщо валюта невідома — повертає None.
    """
    currency_codes = {"USD":840, "EUR":978, "UAH":980}
    return currency_codes.get(currency)
print(get_currency_code("USD"))
````

## 🔹 2. get_exchange_rate(currency_code: int) -> float

Призначення:
* Повертає курс валюти до гривні (UAH) за фіксованою таблицею.

Приклад:

```
get_exchange_rate(840) → 39.5
get_exchange_rate(978) → 42.3
get_exchange_rate(999) → None
```



````python
from typing import Optional
def get_exchange_rate(currency_code: int) -> Optional[float]:
    """
    Приймає числовий код валюти.
    Повертає курс валюти до UAH.
    Якщо код не знайдено — повертає None.
    """
    exchange_rate = {980:1.0, 978:42.3, 840:39.5}
    return exchange_rate.get(currency_code)
print(get_exchange_rate(840))

````

## 🔹 3. get_unique_account_types(accounts: list[dict]) -> set[str]

Призначення:
* Повертає множину унікальних типів рахунків користувача.


Приклад:

`{'fop', 'black', 'white', 'madeInUkraine'}`

````python
def get_unique_account_types(accounts: list[dict]) -> set[str]:
    """
    Приймає список рахунків користувача.
    Повертає множину унікальних значень поля 'type'.
    """
    return {item['type'] for item in accounts if 'type' in item}

print(get_unique_account_types(user_data["accounts"]))
````

## 🔹 4. calculate_total_balance_by_currency(accounts: list[dict]) -> dict[int, int]

Призначення:
* Обчислює сумарний баланс по кожній валюті.


Приклад:

`{980: 19774, 840: 19734, 978: 741}`

````python
def calculate_total_balance_by_currency(accounts: list[dict]) -> dict[int, int]:
    """
    Приймає список рахунків.
    Повертає словник:
        ключ — currencyCode
        значення — сумарний баланс у цій валюті.
    """
    unique_currency_balance = {}
    for item in accounts:
      unique_currency_balance[item['currencyCode']] =(
          unique_currency_balance.get(item['currencyCode'], 0) +
          item['balance']
      )
    return unique_currency_balance
print(calculate_total_balance_by_currency(user_data["accounts"]))
````

## 🔹 5. calculate_total_balance(accounts: list[dict]) -> float

Призначення:
* Обчислює загальний баланс користувача у гривнях (UAH).

````python
def calculate_total_balance(accounts: list[dict]) -> float:
    """
    Приймає список рахунків.
    Конвертує баланс кожного рахунку у гривні
    за допомогою get_exchange_rate().
    Повертає загальну суму у UAH.
    """
    total_uah_balance = 0
    for item in accounts:
      total_uah_balance += item['balance']*get_exchange_rate(item['currencyCode'])
    return round(total_uah_balance / 100, 2)
print(calculate_total_balance(user_data["accounts"]))
````
